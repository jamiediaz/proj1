```python
# Dependencies
%matplotlib notebook
from matplotlib import pyplot as plt
from scipy.stats import linregress
from scipy import stats
import numpy as np
import pandas as pd
```

```python
all_data_path="csv_output/crime_cnt_df.csv"
coldblood_df = pd.read_csv(all_data_path)
coldblood_df.head()
```

```python
avg_temp=coldblood_df["Average Temperature"]
max_temp=coldblood_df["Maximum Temperature"]
crime_rates=coldblood_df["Violent Crime Actual"]
hhincome=coldblood_df["Household Median Income"]
```

```python
# Set data for plotting first question: is there a correlation between avgerage yearly temp in Harris County and the incidence of violent crime?
x_axis = avg_temp
y_axis = crime_rates
```

```python
# Set line showing potential correlation: total violent crime = slope (avg yearly temp)+intercept
(slope, intercept, _, _, _) = linregress(x_axis, y_axis)
fit = slope * x_axis + intercept
```

```python
 # Plotting data
fig, ax = plt.subplots()

fig.suptitle("Average Temp vs Actual Violent Crime in Harris County", fontsize=12, fontweight="bold")

ax.set_xlim(69,74)
ax.set_ylim(26800,36800)

ax.set_xlabel("Avg Yearly Temp (in deg F)")
ax.set_ylabel("Violent Crime Total (Harris County)")

ax.plot(x_axis, y_axis, linewidth=0, marker='o')
ax.plot(x_axis, fit, 'r--')

plt.savefig("../proj1/Avg_TempVCrime.png")
plt.show()
```

```python
#Set samples to run a linear regression for avg yearly temp vs total incidence of violent crime 
s1 = avg_temp
s2 = crime_rates
#run linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(s1,s2)
print(slope,intercept,r_value,p_value, std_err)
slope1=slope
intercept1=intercept
r_value1=r_value
p_value1=p_value
std_err1=std_err
finalstats={"Slope": slope1,
            "Intercept":intercept1,
            "R_value":r_value1,
            "P Value":p_value1,
            "Standard Error":std_err1}
print(finalstats)
```

```python
# Set data for plotting
x_axis2 = max_temp
y_axis2 = crime_rates
```

```python
# Set line
(slope, intercept, _, _, _) = linregress(x_axis2, y_axis2)
fit = slope * x_axis2 + intercept
```

```python
 # Plot data
fig, ax = plt.subplots()

fig.suptitle("Max Temp vs Actual Violent Crime in Harris County", fontsize=12, fontweight="bold")

ax.set_xlim(95,110)
ax.set_ylim(26500,37000)

ax.set_xlabel("Max Yearly Temp (in deg F)")
ax.set_ylabel("Violent Crime Total (Harris County)")

ax.plot(x_axis2, y_axis2, linewidth=0, marker='^')
ax.plot(x_axis2, fit, 'b--')

plt.savefig("../proj1/Max_TempVCrime.png")
plt.show()
```

```python
#Set samples
s3 = max_temp
s2 = crime_rates
#Run Linear Regression to determine line and P value for slope in line total violent crime(t) = slope (maximum yearly temp)+intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(s3,s2)
print(slope, intercept,r_value, p_value, std_err)
slope2=slope
intercept2=intercept
r_value2=r_value
p_value2=p_value
std_err2=std_err
finalstats={"Slope": [slope1,slope2],
            "Intercept":[intercept1,intercept2],
            "R_value":[r_value1,r_value2],
            "P Value":[p_value1,p_value2],
            "Standard Error":[std_err1,std_err2]}
print(finalstats)
```

```python
# Set data for plotting third question: is there a correlation between household income and the incidence of violent crime?
x_axis3 = hhincome
y_axis3 = crime_rates
```

```python
# Set line
(slope, intercept, _, _, _) = linregress(x_axis3, y_axis3)
fit = slope * x_axis3 + intercept
```

```python
 # Plot data median household income and violent crime data for harris county
fig, ax = plt.subplots()

fig.suptitle("Median Household Income vs Actual Violent Crime in Harris County", fontsize=12, fontweight="bold")

ax.set_xlim(38000,60000)
ax.set_ylim(26500,37000)

ax.set_xlabel("Median Household Income $")
ax.set_ylabel("Violent Crime Total (Actual)")

ax.plot(x_axis3, y_axis3, linewidth=0, marker='o')
ax.plot(x_axis3, fit, 'b--')

plt.savefig("../proj1/HHIncome_V_Crime.png")
plt.show()
```

```python
#Set samples
s4 = hhincome
s2 = crime_rates
#Run Linear Regression to determine line and P value for slope in line total violent crime(t) = slope (median household income)+intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(s4,s2)
print(slope, intercept,r_value, p_value, std_err)
slope3=slope
intercept3=intercept
r_value3=r_value
p_value3=p_value
std_err3=std_err
finalstats={"Slope": [slope1,slope2,slope3],
            "Intercept":[intercept1,intercept2,intercept3],
            "R_value":[r_value1,r_value2,r_value3],
            "P Value":[p_value1,p_value2,p_value3],
            "Standard Error":[std_err1,std_err2,std_err3],
            "Line":["yearly violent crime = 310.61(avg yearly temp)+ 9723.62",
                    "yearly violent crime = -137.038(max yearly temp)+45683.02",
                    "yearly violent crime=.028657(median household income)+18068.087"],
           "Significant?": ["No","No","Yes!"]}
print(finalstats)
```

```python
#putting all the final stats from each Linear Regression Analysis into a dataframe
finalstats_df=pd.DataFrame.from_dict(finalstats)
finalstats_df
```
